# Problem 30: Delete Node in a BST
>Given a root node reference of a BST and a key, delete the node with the given key in the BST. Return the root node reference (possibly updated) of the BST.
> Basically, the deletion can be divided into two stages:
>Search for a node to remove.
>If the node is found, delete the node.
>>> 
>>>Input: root = [5,3,6,2,4,null,7], key = 3



>>>Output: [5,4,6,2,null,null,7]
>>>Explanation: Given key to delete is 3. So we find the node with value 3 and delete it.
>>>One valid answer is [5,4,6,2,null,null,7], shown in the above BST.
>>> Please notice that another valid answer is [5,2,6,null,4,null,7] and it's also accepted.
```python
Example 2:
Input: root = [5,3,6,2,4,null,7], key = 0
Output: [5,3,6,2,4,null,7]
Explanation: The tree does not contain a node with value = 0.
Example 3:
Input: root = [], key = 0
Output: []
```

### Define the Binary tree

````python
class TreeNode:
    def __init__(self, value = 0, left = None, right = None):
        self.value = value
        self.left = left
        self.right = right
````

````python
class Solution:
    def DeleteNode(self, root : TreeNode, key : int):
        #if the node is empty
        if not root:
            return None
        
        #if the node is in the left side
        elif key < root.left:
            root.left = self.DeleteNode(root.left, key)
        
        #if the root is in right side
        elif key> root.right:
            root.right = self.DeleteNode(root.right,key)
        
        #now comes the deletion process
        #if the key is equal to the roots value delete the node
        else:
            
            #if the root has no children return none
            if not root.left and not root.right:
                return None
            
            #if the root has one successor in either left or right
            elif not root.right:
                return root.left
            elif not root.left:
                return root.right

            #if the root has two successors, replace it with in order successor
            else:
                 
                 #find the inorder successor, the smallest node is the right subtree
                 succesor = root.right 
                 while succesor.left:
                    succesor = succesor.left
                
                #copy the successor value to the root
                root.val = successor.val

                # Delete the successor from the right subtree
                root.right = self.deleteNode(root.right, successor.val)

        # Return the updated root
        return root

````
