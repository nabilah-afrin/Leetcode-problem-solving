### Iteration Process through bfs
store the nodes in the tree level wise in a queue

pop / enqueue from the queue until a level is fully traversed and while doing so,  add nodes in the next level to the queue.

After every level traversed,  increment the value of levels.

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def maxdepth(root):
    if not root:
        return 0
    q = [root]
    levels = 0

    while q:
        l = len(q)
        for i in range(l):
            current = q.pop(0)
            if current.left:
                q.append(current.left)
            if current.right:
                q.append(current.right)
        levels += 1
    return levels

def create_tree():
    values = input("Enter node values separated by spaces (or -1 for no node): ").split()
    val = int(values[0])
    if val == -1:
        return None
    node = TreeNode(val)
    print("Enter left child of", val)
    node.left = create_tree()
    print("Enter right child of", val)
    node.right = create_tree()
    return node

def main():
    num_test_cases = int(input("Enter the number of test cases: "))
    
    for _ in range(num_test_cases):
        print(f"Test Case {_ + 1}:")
        root = create_tree()
        height = maxdepth(root)
        print("Height of the binary tree:", height)
        print()  # Add a blank line between test cases

if __name__ == "__main__":
    main()
```
